Using Data Science and a k-Means clustering algorithm, we can identify players that are high performing, but may not be as highly paid, that the NBA or Mr. Rooney can target in order to strengthen the team and increase their chances of making the playoffs. 

In order to set up the model, I first combined the performance data from the 2025 NBA season with salary information for each player. The two datasets were merged by player name so that each observation included both how much a player earns in the season and how they performed on the court. I then sorted the dataset based on the number of games played to prioritize players who were consistently active during the season. I also cleaned the data by removing duplicate players, dropping missing values, and selecting only the factors that were most relevant to evaluating player performance, such as their shooting percentages, total points scored, and games played. After standardizing the variables to ensure that they contribute equally and that no single variable dominates the grouping process due to their scale, I performed k-Means clustering. This is a type of clustering algorithm that groups players based on similarities in their performance profiles.

From this model, I was able to identify players who are not good choices, those who are strong choices, and those who could be solid alternatives if the top targets are unavailable for whatever reason. The visualization shows players grouped by performance, with lighter colors representing higher salaries.

Anthony Davis, Bradley Beal, and Trae Young are examples of poor choices based on this analysis. All three have large salaries, yet their standardized performance metrics this past season were below average compared to the rest of the league. Trae Young had the lowest salary of the three at approximately $45 million, but all three showed negative standardized effective field goal percentage (eFG%) and total points scored (PTS). This means that they fall below the midpoint of performance in key categories. When a player is earning this level of financial compensation, but producing a below average performance and score relative to other players, it makes them a high risk investment as it creates a mismatch between cost and production.

On the other hand, Keyonte George, Shaedon Sharpe, and Deni Avdija are considered strong choices because they demonstrate high performance while earning at most 14 million dollars, which is significantly less than the $45 million range of the previous players. The three of them show strong, positive standardized PTS scores, indicating that they perform well above the league average. These are the types of players who outperform their salary tier, providing us with a strong return on investment. By targeting these players, the team can improve their overall efficiency without committing an overwhelming portion of payroll to a single individual - all while increasing their likelihood of making the playoffs.

While they may not rank at the very top of the performance cluster, they are dependable and can get the job done without requiring a massive financial commitment.
